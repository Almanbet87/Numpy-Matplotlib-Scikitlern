```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color = 'g')
plt.show()
```

```python
plt.plot(t, f, color = 'g')
plt.title('График f(t)', fontsize=16, fontweight='bold', color='blue', family='Times', loc='left')
plt.show
```

```python
label_dict = {'fontsize': '18', 'color': 'blue'}

plt.plot(t, f, color = 'g')
plt.title('График f(t)', fontsize=16, fontweight='bold', color='blue', family='Times', loc='left')

plt.xlabel('t', fontdict=label_dict)
plt.ylabel('f', fontdict=label_dict)


plt.show
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])

label_dict = {'fontsize': '18', 'color': 'blue'}

plt.plot(t, f, color = 'g')
plt.title('График f(t)', fontsize=16, fontweight='bold', color='blue', family='Times', loc='left')

plt.xlabel('t', fontdict=label_dict)
plt.ylabel('f', fontdict=label_dict)


plt.show
```
